```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
# FETCHING PAGE CONTENT FUNCTION
def get_page_content(URL):
    
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/12.1.1 Safari/605.1.15'}
    res = requests.get(URL, headers=headers)
    soup = BeautifulSoup(res.content,'html.parser')
    return soup


product_urls = []
source_urls = ['https://www.farfetch.com/fr/shopping/men/shoes-2/items.aspx']
df = pd.DataFrame({'BRAND_NAME':[],'PRODUCT_NAME':[],'PRODUCT_CATEGORY':[],'ORIGINAL_PRICE':[],'SELLING_PRICE':[],'DISCOUNT':[],'IMAGE_URLS':[],'PRODUCT_URL':[]})
for lin in source_urls:
    page_num = 1
    # EXTRACTING ALL THE PAGINATION URLS
    while True:
        pagination_urls = []
        base_url = f'{lin}?page={page_num}'
        soup = get_page_content(base_url)
        href_links = soup.find('ul','_259180').findAll('a','css-n5nq0d-ProductCardLink e4l1wga0')
        domain = "https://www.farfetch.com"
        print(f"EXTRACTING {base_url} ..........\n\n\n")
        for link in href_links:
            url = domain + link.get('href')
            pagination_urls.append(url)

        # USING SET FUNCTION TO  COMPARE THE PRODUCT URLS, IF THE PAGINATION EXCEEDED IT REDIRECTING TO THE FIRST PAGE  
        if len(set(product_urls) & set(pagination_urls)) >= 5:
            break
        product_urls.extend(pagination_urls)
        page_num += 1
    print('LOOP BREAKED')
    

# EXTRACTING DATA FROM THE INDIVIDUAL PRODUCT URLS
for url in product_urls:
    print(f"EXTRACTING {url} ..........")
    soup = get_page_content(url)
    if soup.find('ol','_bb8ef4') == None:
        continue
    Category = soup.find('ol','_bb8ef4').findAll('li','_84ddad')[2].text
    Price_Discount = soup.find('div','_7ce535').findAll('span')
    if len(Price_Discount) >= 2:
        original_price = soup.find('div','_7ce535').find('span','_864ecb _1fe24a').text
        sell_price = soup.find('div','_7ce535').find('strong','_2f9add _2231e7 _b4693b').text
        discount = soup.find('div','_7ce535').find('span','_121750').text
        original_price = re.sub("\D", "", original_price)
        sell_price = re.sub("\D", "", sell_price)
        discount = re.sub("\D", "", discount)
    if len(Price_Discount) == 1:
        original_price = soup.find('div','_7ce535').find('span','_2f9add _29273d').text
        sell_price = 0
        discount = 0
        original_price = re.sub("\D", "", original_price)
    brand_name = soup.find('h1','_6b7f2f').find('span','_af2f5c').text
    product_name = soup.find('h1','_6b7f2f').find('span','_b4693b _2ef212 _b4693b').text
    image_tab = soup.find('div','_4accab').findAll('link')
    image_href = image_tab[0].get('href'),image_tab[1].get('href')
    
    # STORING ALL THE REQUIRD DATA IN TEMPORARY PAYLOAD AND PRINTING IT
    payload = {
        "BRAND_NAME" : brand_name,
        "PRODUCT_NAME" : product_name,
        "PRODUCT_CATEGORY" : Category,
        "ORIGINAL_PRICE" : original_price,
        "SELLING_PRICE" : sell_price,
        "DISCOUNT" : discount,
        "IMAGE_URLS" : image_href,
        "PRODUCT_URL" : url
    }
    print(f"\n{payload}\n\n")
    df = df.append(payload, ignore_index = True)
```
